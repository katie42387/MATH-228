```python
import networkx as nx
G = nx.Graph()
```

```python
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from( [(1, 2), (1, 3), (2, 4), (2, 3), (3, 4), (1, 4)])
```

```python
G.number_of_nodes(), G.number_of_edges()
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G)
```

```python
nx.draw_networkx(G, with_labels=True)
```

```python
nx.triangles(G)
```

```python
nx.draw_networkx(nx.minimum_spanning_tree(G), with_labels=True)
```
