### Chromatic Number

What we are really interested in is, given a graph $G$, what is the minimal number of colors needed to give a valid coloring. (you might recall from class that once we have a valid color with $n$ colors, it is easy to add more colors). 

The **Chromatic Number** of a graph G, $\chi(G)$ is the minimum number of colors necessary to achieve a valid coloring.

For small examples this is easy enough. Computer $\chi(G)$ for the following three graphs:

#### Q1: Find the chromatic number of the complete graph on 6 vertices

```python
G = nx.Graph()

# A little python trick for getting the integers from 0 ... 5.
G.add_nodes_from(range(0, 6))

# Here is a little python trick for getting all of the pairs of numbers
# from the integers from 0 ... 5
edges = [ (x, y) for x in range(0, 6) for y in range(x, 6)]
G.add_edges_from(edges)

nx.draw(G)
```

#### Q2: Find the chromatic number of the graph six vertices below

```python
G = nx.Graph()

# A little python trick for getting the integers from 0 ... 5.
G.add_nodes_from(range(0, 6))

G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (1, 4), (2, 4), (2, 5),
                 (3, 4), (4, 5)])

nx.draw(G)
```

#### Q3: Find the chromatic number of the graph with five vertices below

```python
G = nx.Graph()

# A little python trick for getting the integers from 0 ... 5.
G.add_nodes_from(range(0, 5))

G.add_edges_from([(0, 3), (0, 4), (1, 3), (1, 4), (2, 3), (2, 4)])

nx.draw(G)
```

## The Four Color Theorem

The four color theorem is that any planar graph $G$, has $\chi(G) \leq 4$. I.e. at most 4 colors are needed.

In 1976 this theorem was proven by Appel and Haken. It is the first example of a proof that **required** a computer, i.e. that could not have been done without the computer. There method was to first show that if there was a counter example to the theorem then it had to come from one of 1834 possible configurations. They then used a computer to show that for all of these 1834 possible counter examples a 4 coloring (or better) was possible. Note why they needed a computer. Even if each of them had checked one of these possible countetr examples per day it would have taken them 2.5 years to check them all.

This proof is not without controversy. As it was the first example of a proof that would not be human verifiable. I sort of object to this characterization, as it would be humanverifiable it would just take an extraordinary effort.

In 2005, Gonthier proved the theorem using an automated theorem prover (more about these later in this class) and produced a result that the prover found the theorem was true. The resulting series of logical statements though is so long that no one has been able to go through and verify each line. Meaning that the theorem still remains not verified by a human.
